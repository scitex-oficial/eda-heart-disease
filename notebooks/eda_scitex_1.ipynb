> **Fuente:** UCI Machine Learning Repository · Heart Disease (Cleveland) · id=45

```python
from ucimlrepo import fetch_ucirepo

heart = fetch_ucirepo(id=45)
X = heart.data.features
y = heart.data.targets

df = pd.concat([X, y], axis=1)
df['target'] = (df['num'] > 0).astype(int)
df = df.drop(columns='num')

print(f"✓ Dataset cargado: {df.shape[0]} filas · {df.shape[1]} columnas")
print(f"  Sin enfermedad : {(df['target']==0).sum()} pacientes")
print(f"  Con enfermedad : {(df['target']==1).sum()} pacientes")
print(f"  Valores nulos  :\n{df.isnull().sum()[df.isnull().sum()>0]}")
```

```python
# ── Paleta SciTeX ──────────────────────────────────────────────────────────────
COLOR_NAVY   = '#27295a'
COLOR_ACCENT = '#3c78bb'
COLOR_GRAY   = '#71706f'
COLOR_BG     = '#f8f9fb'
COLOR_RED    = '#993C1D'

# ── Configuración global ───────────────────────────────────────────────────────
sns.set_theme(style='whitegrid', font='DejaVu Sans')
plt.rcParams.update({
    'axes.facecolor':    COLOR_BG,
    'figure.facecolor':  COLOR_BG,
    'axes.edgecolor':    '#ccccdd',
    'axes.labelcolor':   COLOR_NAVY,
    'xtick.color':       COLOR_GRAY,
    'ytick.color':       COLOR_GRAY,
    'axes.spines.top':   False,
    'axes.spines.right': False,
    'grid.color':        '#e0e0ec',
    'grid.linewidth':    0.6,
    'svg.fonttype':      'none',
})

# ── Preparar data ──────────────────────────────────────────────────────────────
target_map = {0: 'Sin enfermedad', 1: 'Con enfermedad'}
df['target_label'] = df['target'].map(target_map)

VAR   = 'thalach'
LABEL = 'Frecuencia Cardíaca Máxima (lpm)'

n_sin   = df[df['target_label'] == 'Sin enfermedad'].shape[0]
n_con   = df[df['target_label'] == 'Con enfermedad'].shape[0]
med_sin = df[df['target_label'] == 'Sin enfermedad'][VAR].median()
med_con = df[df['target_label'] == 'Con enfermedad'][VAR].median()

print(f"✓ Configuración aplicada")
print(f"  Sin enfermedad → n={n_sin} · mediana={med_sin:.0f} lpm")
print(f"  Con enfermedad → n={n_con} · mediana={med_con:.0f} lpm")
print(f"  Diferencia de medianas: {med_sin - med_con:.0f} lpm")
```

<div style="margin: 24px 0 8px 0;">

```python
resumen = df.groupby('target_label')[VAR].agg(
    n='count',
    media='mean',
    mediana='median',
    desv_std='std',
    minimo='min',
    maximo='max'
).round(1)

resumen.index.name = 'Diagnóstico'
resumen.columns = ['N', 'Media', 'Mediana', 'Desv. Std', 'Mínimo', 'Máximo']

resumen.style\
    .set_caption('Frecuencia Cardíaca Máxima (lpm) por Diagnóstico')\
    .set_table_styles([
        {'selector': 'caption',
         'props': [('font-size', '13px'), ('font-weight', 'bold'),
                   ('color', '#27295a'), ('padding-bottom', '8px')]},
        {'selector': 'th',
         'props': [('background-color', '#27295a'), ('color', 'white'),
                   ('font-size', '12px'), ('padding', '8px 14px')]},
        {'selector': 'td',
         'props': [('font-size', '12px'), ('padding', '6px 14px'),
                   ('color', '#27295a')]},
        {'selector': 'tr:nth-child(even)',
         'props': [('background-color', '#f0f0f8')]},
        {'selector': 'tr:hover',
         'props': [('background-color', '#e0e8f5')]},
    ])\
    .highlight_max(color='#d4edda')\
    .highlight_min(color='#f8d7da')
```

  <strong style="color: #27295a;">Insight:</strong>
  Los pacientes con enfermedad cardíaca alcanzan una frecuencia cardíaca
  máxima notablemente más baja. La diferencia de medianas es de
  <strong style="color: #993C1D;">~19 lpm</strong>, un discriminador
  clínico relevante.
</div>

```python
fig, ax = plt.subplots(figsize=(8.4, 8.6))
fig.patch.set_facecolor(COLOR_BG)

for tval, color in [
    ('Sin enfermedad', COLOR_RED),
    ('Con enfermedad', COLOR_ACCENT),
]:
    sns.histplot(
        data=df[df['target_label'] == tval],
        x=VAR,
        color=color,
        kde=True,
        bins=22,
        alpha=0.45,
        linewidth=0.4,
        line_kws={'linewidth': 2.5},
        ax=ax,
    )

ymax = ax.get_ylim()[1]
for tval, color in [
    ('Sin enfermedad', COLOR_RED),
    ('Con enfermedad', COLOR_ACCENT),
]:
    med    = df[df['target_label'] == tval][VAR].median()
    offset = +2 if tval == 'Sin enfermedad' else -18
    ax.axvline(med, color=color, linewidth=2, linestyle='--', alpha=0.85)
    ax.text(
        med + offset, ymax * 0.80,
        f'Mediana\n{med:.0f} lpm',
        color=color, fontsize=9, fontweight='bold', va='top',
        bbox=dict(boxstyle='round,pad=0.35', facecolor='white',
                  edgecolor=color, alpha=0.85)
    )

ax.legend(
    handles=[
        mpatches.Patch(facecolor=COLOR_RED, alpha=0.6,
                       label=f'Sin enfermedad  (n={n_sin})'),
        mpatches.Patch(facecolor=COLOR_ACCENT, alpha=0.6,
                       label=f'Con enfermedad  (n={n_con})'),
    ],
    title='Diagnóstico', title_fontsize=11,
    fontsize=11, framealpha=0.9,
    edgecolor='#ccccdd', loc='upper left',
)
ax.set_title(
    'Frecuencia Cardíaca Máxima por Diagnóstico',
    fontsize=15, fontweight='bold', color=COLOR_NAVY, pad=14
)
ax.set_xlabel(LABEL, fontsize=12, color=COLOR_GRAY, labelpad=8)
ax.set_ylabel('Frecuencia', fontsize=12, color=COLOR_GRAY)
ax.yaxis.grid(True)
ax.set_axisbelow(True)
ax.text(
    0.98, 0.97,
    f'Diferencia de medianas: {med_sin - med_con:.0f} lpm\n'
    'Pacientes con enfermedad alcanzan\nuna FC máxima más baja.',
    transform=ax.transAxes, fontsize=9.5, va='top', ha='right',
    bbox=dict(boxstyle='round,pad=0.5', facecolor='white',
              edgecolor=COLOR_ACCENT, alpha=0.92),
    color=COLOR_NAVY
)

fig.text(0.02, 0.005, 'Dataset: Heart Disease UCI · n=303',
         fontsize=8, color=COLOR_GRAY)
fig.text(0.98, 0.005, '#PensarConDatos',
         fontsize=8, color=COLOR_ACCENT, ha='right')

plt.tight_layout(rect=[0, 0.02, 1, 1])
plt.savefig('thalach_histplot.svg',
            bbox_inches='tight', facecolor=COLOR_BG, format='svg')
plt.savefig('thalach_histplot.png',
            dpi=150, bbox_inches='tight', facecolor=COLOR_BG)
print("✓ Gráfico exportado: thalach_histplot.svg / .png")
plt.show()
```

```python
VAR   = 'thalach'
LABEL = 'Frecuencia Cardíaca Máxima (lpm)'

n_sin   = df[df['target_label'] == 'Sin enfermedad'].shape[0]
n_con   = df[df['target_label'] == 'Con enfermedad'].shape[0]
med_sin = df[df['target_label'] == 'Sin enfermedad'][VAR].median()
med_con = df[df['target_label'] == 'Con enfermedad'][VAR].median()

fig, ax = plt.subplots(figsize=(8.4, 8.6))
fig.patch.set_facecolor(COLOR_BG)

for tval, color in [
    ('Sin enfermedad', COLOR_RED),
    ('Con enfermedad', COLOR_ACCENT),
]:
    sns.kdeplot(
        df[df['target_label'] == tval][VAR],
        ax=ax, color=color, linewidth=2.5,
        fill=True, alpha=0.22,
    )

ymax = ax.get_ylim()[1]
for tval, color in [
    ('Sin enfermedad', COLOR_RED),
    ('Con enfermedad', COLOR_ACCENT),
]:
    med    = df[df['target_label'] == tval][VAR].median()
    offset = +2 if tval == 'Sin enfermedad' else -18
    ax.axvline(med, color=color, linewidth=2, linestyle='--', alpha=0.85)
    ax.text(
        med + offset, ymax * 0.80,
        f'Mediana\n{med:.0f} lpm',
        color=color, fontsize=9, fontweight='bold', va='top',
        bbox=dict(boxstyle='round,pad=0.35', facecolor='white',
                  edgecolor=color, alpha=0.85)
    )

ax.axvspan(med_con, med_sin, alpha=0.06, color=COLOR_NAVY)

ax.legend(
    handles=[
        mpatches.Patch(facecolor=COLOR_RED, alpha=0.5,
                       label=f'Sin enfermedad  (n={n_sin})'),
        mpatches.Patch(facecolor=COLOR_ACCENT, alpha=0.5,
                       label=f'Con enfermedad  (n={n_con})'),
        mpatches.Patch(facecolor=COLOR_NAVY, alpha=0.12,
                       label=f'Diferencia de medianas: {med_sin - med_con:.0f} lpm'),
    ],
    title='Diagnóstico', title_fontsize=11,
    fontsize=11, framealpha=0.9,
    edgecolor='#ccccdd', loc='upper left',
)
ax.set_title(
    'Frecuencia Cardíaca Máxima por Diagnóstico\n(curva de densidad)',
    fontsize=15, fontweight='bold', color=COLOR_NAVY, pad=14
)
ax.set_xlabel(LABEL, fontsize=12, color=COLOR_GRAY, labelpad=8)
ax.set_ylabel('Densidad', fontsize=12, color=COLOR_GRAY)
ax.yaxis.grid(True)
ax.set_axisbelow(True)
ax.text(
    0.98, 0.97,
    f'Diferencia de medianas: {med_sin - med_con:.0f} lpm\n'
    'La FC máxima es un discriminador\nclave entre ambos grupos.',
    transform=ax.transAxes, fontsize=9.5, va='top', ha='right',
    bbox=dict(boxstyle='round,pad=0.5', facecolor='white',
              edgecolor=COLOR_ACCENT, alpha=0.92),
    color=COLOR_NAVY
)
fig.text(0.02, 0.005, 'Dataset: Heart Disease UCI · n=303',
         fontsize=8, color=COLOR_GRAY)
fig.text(0.98, 0.005, '#PensarConDatos',
         fontsize=8, color=COLOR_ACCENT, ha='right')

plt.tight_layout(rect=[0, 0.02, 1, 1])
plt.savefig('thalach_kdeplot.svg', bbox_inches='tight',
            facecolor=COLOR_BG, format='svg')
plt.savefig('thalach_kdeplot.png', dpi=150,
            bbox_inches='tight', facecolor=COLOR_BG)
print("✓ thalach_kdeplot.svg / .png")
plt.show()
```

```python
vars_config = [
    ('age',      'Edad (años)',                      'upper right', +0.5, -6,   '{:.0f} años'),
    ('thalach',  'FC Máxima (lpm)',                  'upper left',  +2,   -18,  '{:.0f} lpm'),
    ('trestbps', 'Presión Arterial (mmHg)',           'upper right', +1,   -12,  '{:.0f} mmHg'),
    ('chol',     'Colesterol (mg/dl)',                'upper right', +2,   -22,  '{:.0f} mg/dl'),
    ('oldpeak',  'Depresión ST (oldpeak)',            'upper right', +0.1, -0.9, '{:.1f}'),
]

fig, axes = plt.subplots(2, 3, figsize=(18, 12))
fig.patch.set_facecolor(COLOR_BG)
axes_flat = axes.flat

for ax, (var, label, leg_loc, off_sin, off_con, fmt) in zip(axes_flat, vars_config):

    for tval, color in [
        ('Sin enfermedad', COLOR_RED),
        ('Con enfermedad', COLOR_ACCENT),
    ]:
        sns.histplot(
            data=df[df['target_label'] == tval],
            x=var, color=color, kde=True,
            bins=20, alpha=0.45, linewidth=0.4,
            line_kws={'linewidth': 2.0}, ax=ax,
        )

    ymax = ax.get_ylim()[1]
    for tval, color, offset in [
        ('Sin enfermedad', COLOR_RED,    off_sin),
        ('Con enfermedad', COLOR_ACCENT, off_con),
    ]:
        med = df[df['target_label'] == tval][var].median()
        ax.axvline(med, color=color, linewidth=1.8,
                   linestyle='--', alpha=0.85)
        ax.text(
            med + offset, ymax * 0.78,
            f'Med.\n{fmt.format(med)}',
            color=color, fontsize=8, fontweight='bold', va='top',
            bbox=dict(boxstyle='round,pad=0.25', facecolor='white',
                      edgecolor=color, alpha=0.85)
        )

    med_sin = df[df['target_label'] == 'Sin enfermedad'][var].median()
    med_con = df[df['target_label'] == 'Con enfermedad'][var].median()

    ax.legend(
        handles=[
            mpatches.Patch(facecolor=COLOR_RED, alpha=0.6,
                           label=f'Sin enfermedad (n={n_sin})'),
            mpatches.Patch(facecolor=COLOR_ACCENT, alpha=0.6,
                           label=f'Con enfermedad (n={n_con})'),
        ],
        title='Diagnóstico', title_fontsize=8,
        fontsize=8, framealpha=0.9,
        edgecolor='#ccccdd', loc=leg_loc,
    )
    ax.set_title(label, fontsize=12, fontweight='bold',
                 color=COLOR_NAVY, pad=8)
    ax.set_xlabel('')
    ax.set_ylabel('Frecuencia', fontsize=9, color=COLOR_GRAY)
    ax.yaxis.grid(True)
    ax.set_axisbelow(True)

# Ocultar subplot vacío
axes_flat[5].set_visible(False)

fig.suptitle(
    'Variables Cuantitativas por Diagnóstico Cardíaco Histograma + Curva de Densidad',
    fontsize=16, fontweight='bold', color=COLOR_NAVY, y=1.01
)
fig.text(0.5, 0.98,
         'Dataset: Heart Disease UCI · n=303  |  Línea punteada = mediana por grupo',
         ha='center', fontsize=9, color=COLOR_GRAY)
fig.text(0.02, 0.005, 'MANUEL MERINO · by SciTeX',
         fontsize=8, color=COLOR_NAVY, alpha=0.5, fontweight='bold')
fig.text(0.98, 0.005, '#PensarConDatos',
         fontsize=8, color=COLOR_ACCENT, ha='right')

plt.tight_layout(rect=[0, 0.02, 1, 0.98])
plt.savefig('panel_histplot.svg', bbox_inches='tight',
            facecolor=COLOR_BG, format='svg')
plt.savefig('panel_histplot.png', dpi=150,
            bbox_inches='tight', facecolor=COLOR_BG)
print("✓ panel_histplot.svg / .png")
plt.show()
```

<div style="background: #f0f0f8; border-left: 4px solid #3c78bb;
            padding: 10px 16px; border-radius: 0 6px 6px 0;
            font-family: sans-serif; font-size: 12px; color: #71706f;
            margin-bottom: 8px;">
  <strong style="color: #27295a;">Insight:</strong>
  El kdeplot elimina el ruido visual de las barras. El área sombreada
  entre medianas cuantifica la separación real entre grupos.
  Compara con el panel anterior — la forma de las distribuciones
  queda más clara sin las barras.
</div>

```python
fig, axes = plt.subplots(2, 3, figsize=(18, 12))
fig.patch.set_facecolor(COLOR_BG)
axes_flat = axes.flat

for ax, (var, label, leg_loc, off_sin, off_con, fmt) in zip(axes_flat, vars_config):

    for tval, color in [
        ('Sin enfermedad', COLOR_RED),
        ('Con enfermedad', COLOR_ACCENT),
    ]:
        sns.kdeplot(
            df[df['target_label'] == tval][var],
            ax=ax, color=color, linewidth=2.2,
            fill=True, alpha=0.22,
        )

    ymax = ax.get_ylim()[1]
    med_sin = df[df['target_label'] == 'Sin enfermedad'][var].median()
    med_con = df[df['target_label'] == 'Con enfermedad'][var].median()

    for tval, color, offset in [
        ('Sin enfermedad', COLOR_RED,    off_sin),
        ('Con enfermedad', COLOR_ACCENT, off_con),
    ]:
        med = df[df['target_label'] == tval][var].median()
        ax.axvline(med, color=color, linewidth=1.8,
                   linestyle='--', alpha=0.85)
        ax.text(
            med + offset, ymax * 0.78,
            f'Med.\n{fmt.format(med)}',
            color=color, fontsize=8, fontweight='bold', va='top',
            bbox=dict(boxstyle='round,pad=0.25', facecolor='white',
                      edgecolor=color, alpha=0.85)
        )

    ax.axvspan(
        min(med_sin, med_con),
        max(med_sin, med_con),
        alpha=0.06, color=COLOR_NAVY
    )

    ax.legend(
        handles=[
            mpatches.Patch(facecolor=COLOR_RED, alpha=0.5,
                           label=f'Sin enfermedad (n={n_sin})'),
            mpatches.Patch(facecolor=COLOR_ACCENT, alpha=0.5,
                           label=f'Con enfermedad (n={n_con})'),
            mpatches.Patch(facecolor=COLOR_NAVY, alpha=0.12,
                           label=f'Δ med: {abs(med_con - med_sin):.1f}'),
        ],
        title='Diagnóstico', title_fontsize=8,
        fontsize=8, framealpha=0.9,
        edgecolor='#ccccdd', loc=leg_loc,
    )
    ax.set_title(label, fontsize=12, fontweight='bold',
                 color=COLOR_NAVY, pad=8)
    ax.set_xlabel('')
    ax.set_ylabel('Densidad', fontsize=9, color=COLOR_GRAY)
    ax.yaxis.grid(True)
    ax.set_axisbelow(True)

axes_flat[5].set_visible(False)

fig.suptitle(
    'Variables Cuantitativas por Diagnóstico Cardíaco + Curva de Densidad (KDE)',
    fontsize=16, fontweight='bold', color=COLOR_NAVY, y=1.01
)
fig.text(0.5, 0.98,
         'Dataset: Heart Disease UCI · n=303  |  '
         'Área sombreada = diferencia entre medianas',
         ha='center', fontsize=9, color=COLOR_GRAY)
fig.text(0.02, 0.005, 'MANUEL MERINO · by SciTeX',
         fontsize=8, color=COLOR_NAVY, alpha=0.5, fontweight='bold')
fig.text(0.98, 0.005, '#PensarConDatos',
         fontsize=8, color=COLOR_ACCENT, ha='right')

plt.tight_layout(rect=[0, 0.02, 1, 0.98])
plt.savefig('panel_kdeplot.svg', bbox_inches='tight',
            facecolor=COLOR_BG, format='svg')
plt.savefig('panel_kdeplot.png', dpi=150,
            bbox_inches='tight', facecolor=COLOR_BG)
print("✓ panel_kdeplot.svg / .png")
plt.show()
```
